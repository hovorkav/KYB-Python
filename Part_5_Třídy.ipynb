# Třídy

(zjednodušený přehled)

Objekty jsou v podstatě soubory proměnných (jedna nebo více) a jejich chování (funkce).

### Příklad třídy obsahující pouze atribut (proměnnou), bez funkcí:

```python
class Osoba:
    def __init__(self, jmeno="defaultni_jmeno"):
        self.jmeno = jmeno  # atribut instance, s výchozí hodnotou

# vytvoření instancí třídy
x = Osoba()           # instance s výchozím jménem
y = Osoba("Jana")     # instance s vlastním jménem

print(x.jmeno)  # vypíše "defaultni_jmeno"
print(y.jmeno)  # vypíše "Jana"

x.jmeno = "Karel"     # změna atributu jmeno pro instanci x
print(x.jmeno)        # vypíše "Karel"
print(y.jmeno)        # stále vypíše "Jana"
```

### Příklad třídy, která obsahuje atributy a funkce (metody):

```python
class Ctverec:
    def __init__(self, strana=0):
        self.strana = strana

    def spocitej_obvod(self):
        return 4 * self.strana

    def spocitej_obsah(self):
        return self.strana ** 2

ctv1 = Ctverec(5)  # vytvoření nové instance třídy Ctverec s délkou strany 5
print(ctv1.spocitej_obvod())  # vypíše 20
print(ctv1.spocitej_obsah())  # vypíše 25
```
